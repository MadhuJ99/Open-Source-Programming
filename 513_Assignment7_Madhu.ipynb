(INDEXING AND SLICING ARRAYS) Create an array containing the values 1–15, reshape it
into a 3-by-5 array, then use indexing and slicing techniques to perform each of the following
operations:
a) Select row 2.
b) Select column 4.
c) Select rows 0 and 1.
d) Select columns 2–4.
e) Select the element that is in row 1 and column 4.
f) Select all elements from rows 1 and 2 that are in columns 0, 2 and 4.

Name: Madhu

```python
import numpy as np

# An array containing the values 1–15,
arraylist = np.arange(1,16)
print("Array list is: ", arraylist)

# reshape array into 3x5 array
finallist = np.reshape(arraylist,(3,5))
print("\n Reshaped array is: \n", finallist)

# a) select row 2
print("\n row 2 is : ", end='')
print(finallist[2])

# b) select column 4
print("\n column 4 is : \n", end='')
print(finallist[0:3, 3:4])

# c) select rows 0 and 1
print("\n rows 0 and 1 : \n", end='')
print(finallist[0:2])

# d) select columns 2 - 4
print("\n column 2 - 4 : \n", end='')
print(finallist[0:3, 2:5])

# e) Select the element that is in row 1 and column 4.
print("\n The element that is in row 1 and column 4: \n ", end='')
print(finallist[1:2, 4:5])


# f) select element in rows 1 and 2 that are in columns 0, 2 and 4
print("\nElement at row 1 and 2 that are in column 0: ", end='')
print(finallist[1:3, 0])
print("Element at row 1 and 2 that are in column 2: ", end='')
print(finallist[1:3, 2])
print("Element at row 1 and 2 that are in column 4: ", end='')
print(finallist[1:3, 4])
```

Write a program to load a given dataset called Student_Grades.csv into a NumPy array. Then
determine the following items:
a) Display all data on screen.
b) Determine how many students were in the dataset?
c) Display the number of rows and columns of your numpy array.
d) Display the array data types.
e) Display the following Descriptive Statistics of students’ overall percentage scores:
a. Min score
b. Max score
c. Mean value
d. Median
e. Mode
f. Standard Deviation
g. 25% and 75% percentile
f) Determine how many students achieved an A grade, B, C, D and F grades.
g) Create a pie chart based on the above grade achievements (option f)

```python
import numpy as np
import pandas as pd
from IPython.display import display, HTML


#a) Display all data on screen.

StudentData = pd.read_csv('Student_Grades.csv')
cols=StudentData.columns
# print out and display dataframe as tables in HTML
display(HTML(StudentData.to_html()))
```

```python
# b) Determine how many students were in the dataset? 

index = StudentData.index
Total_students = len(index)
print("Total number of students are :", Total_students)
```

```python
# c) Display the number of rows and columns of your numpy array. 

Total_data = StudentData.shape
print("Number of rows and columns : ", Total_data)
```

```python
# d) Display the array data types.


Array_data_types = print('Column Datatypes:\n\n', StudentData.dtypes)
```

```python
# e) Display the following Descriptive Statistics of students’ overall percentage scores: a. Min score b. Max score 
# c. Mean value d. Median e. Mode f. Standard Deviation g. 25% and 75% percentile 

#print(StudentData[1:32])

overall_percentage_scores = StudentData.iloc[: , -1]
#print(overall_percentage_scores)


print("Min score is : ", overall_percentage_scores.min())

print("Max score is : ", overall_percentage_scores.max())

print("Mean value is : ", overall_percentage_scores.mean())

print("Median value is : ", overall_percentage_scores.median())

print("Mode is : \n" , overall_percentage_scores.mode())

print("Standard deviation is : ", overall_percentage_scores.std())

print("25th percentile of percentage scores : ", np.percentile(overall_percentage_scores, 25))
print("75th percentile of percentage scores : ", np.percentile(overall_percentage_scores, 75))
```

```python
#  f) Determine how many students achieved an A grade, B, C, D and F grades.


A = print("Number of students who got A grade : ",overall_percentage_scores[overall_percentage_scores >= 90].count())

B = print("Number of students who got B grade : ",overall_percentage_scores[(90 > overall_percentage_scores)
                                                                        &(overall_percentage_scores >= 80)].count())

C = print("Number of students who got C grade : ",overall_percentage_scores[(80 > overall_percentage_scores)
                                                                        &(overall_percentage_scores >= 70)].count())

D = print("Number of students who got D grade : ",overall_percentage_scores[(70 > overall_percentage_scores)
                                                                        &(overall_percentage_scores >= 60)].count())

F = print("Number of students who got F grade : ",overall_percentage_scores[overall_percentage_scores <= 60].count())

```

```python
# g) Create a pie chart based on the above grade achievements (option f)

import matplotlib.pyplot as plt

y = np.array([21, 8, 1, 1, 2])
mylabels = ['A Grade students', 'B Grade students', 'C Grade students', 'D Grade students', 'F Grade students']

plt.figure(figsize=(8,6))
plt.pie(y, labels = mylabels, autopct='%1.1f%%')
plt.title("Students per Grade")
plt.show()

```
