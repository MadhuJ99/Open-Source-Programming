```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML

# a) Using dataset workerstips.csv, create a scatterplot with total_bill as the x-axis and the tips as the y-axis.


Workerstips = pd.read_csv('workerstips.csv')
cols=Workerstips.columns
# print out and display dataframe as tables in HTML
display(HTML(Workerstips.head(5).to_html()))


sns.scatterplot(x="total_bill", y="tip", data=Workerstips)

plt.show()                                                                  # Show plot
```

```python
# b) Using dataset workerstips.csv, create a scatter plot that will differentiate between smokers and non-smokers. 
# Use a different color (hue) and style marker (style) to visually split the points of scatterplot. Because the default 
# range sizes are limited, increase the range size.Try the tuple (10, 300) as the size range parameter.
#sns.set(rc = {'figure.figsize':(10,300)})


sns.scatterplot(x="total_bill", y="tip", hue="smoker", marker='o', sizes=(10, 300), data=Workerstips);
```

```python
# c) A simple bar plot will work well to show average tips per day of the week. Find out which day has the highest average tip.
# In this case, you may want to show the average tip (y-axis) per day of the week (x-axis).

sns.set(rc = {'figure.figsize':(7,7)})
sns.barplot(x="day", y="tip", data=Workerstips);

print("\n As per the below graph Sunday has the highest average tip.")
```

```python
# d) Repeat part (b) above but this time split the distribution by lunch and dinner. What information can you derive from 
# the new visualization?
ax1 = sns.boxplot(x="day", y="total_bill", hue="time", data=Workerstips, linewidth=2.5)


print("""\n From the new visualization I can say people tends to have a dinner more than a lunch outside.
People eat dinner outside on Sunday, Saturday and Friday. Workers have lunch outside on Thurday and Friday only. """ )
```

```python
# d) Repeat part (b) above but this time split the distribution by lunch and dinner. What information can you derive from 
# the new visualization?
ax1 = sns.scatterplot(x="total_bill", y="tip", hue="time", marker = 'o', data=Workerstips)


print("""\n From the new visualization I can say people tends to have a dinner more than a lunch outside. """ )
```

```python
# e) Using dataset flightData.csv, create a lineplot to plot average number of passengers per year per month. 
# Use the year on the x-axis and the passengers on the y-axis. 

flightData = pd.read_csv('flightsData.csv')
cols=flightData.columns
# print out and display dataframe as tables in HTML
display(HTML(flightData.head(5).to_html()))


sns.relplot(data=flightData, x="year", y="passengers", hue="month", kind="line")

```

```python
# Find out how to plot the total number of passengers per year. 


total_passengers = flightData.pivot(index="year", columns="month", values="passengers")
total_passengers.head()

yearly_passengers = flightData.groupby('year').sum().reset_index()
print("The number of passengers per year:  \n\n",yearly_passengers)


sns.barplot(data=yearly_passengers, x="year", y="passengers")
```

```python
# f) Using dataset titanic.csv, show the counts of observations in each categorical bin using bars plot.


TitanicData = pd.read_csv('titanic.csv')
cols=TitanicData.columns
# print out and display dataframe as tables in HTML
display(HTML(TitanicData.head(5).to_html()))
```

```python
TitanicData.groupby(['Sex','Survived'])['Survived'].count()
f,ax=plt.subplots(1,3,figsize=(10,8))
TitanicData[['Sex','Survived']].groupby(['Sex']).mean().plot.bar(ax=ax[0])
ax[0].set_title('Survived vs Sex')
sns.countplot('Sex',hue='Survived',data=TitanicData,ax=ax[1])
ax[1].set_title('Survived vs Dead')
TitanicData.groupby(['Pclass','Sex'])['Sex'].count()
sns.countplot('Pclass',hue='Survived',data=TitanicData,ax=ax[2])
ax[2].set_title('Survived with the class')
plt.show()



print(""" As per the above bar plots the number of male were more than number of female on the ship. But in the number of 
people saved the count of female was almost twice the male count. The survival rate of female was around 75% and the survival
rate of male wass around 18-19%. Survived count of first class is more than second and third class.""")
```
