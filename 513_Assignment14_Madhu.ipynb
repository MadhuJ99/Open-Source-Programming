```python
# c. Use an INNER JOIN to select all books for a specific author. Include the title, copyright year and ISBN. 
# Order the information alphabetically by title.

pd.read_sql("""SELECT authors.first, authors.last, titles.title, titles.copyright, author_ISBN.isbn
            FROM ((authors 
            INNER JOIN author_ISBN ON authors.id = author_ISBN.id)
            INNER JOIN titles ON author_ISBN.isbn = titles.isbn)""", connection).head(10)
```

```python
# d. Insert a new author into a authors table.

cursor = connection.cursor()

cursor = cursor.execute(""" INSERT INTO authors (first, last)
                         VALUES ('Madhu', 'Jagdale')""")
pd.read_sql('SELECT * FROM authors', connection, index_col = ['id'])
```

Open the books database and use cursor method execute to select all the data in the titles table then use description
and fetchall to display the data in tabular format.

Name: Madhu

```python
cursor = connection.cursor()
cursor.execute('SELECT * FROM titles')
# get all records
records = cursor.fetchall()
print("Total number of rows in table: ", cursor.description)
print("\nPrinting each row")
for row in records:
    print("isbn = ", row[0], )
    print("title = ", row[1])
    print("edition  = ", row[2])
    print("copyright  = ", row[3], "\n")
```

```python
from tabulate import tabulate

d = cursor.execute('SELECT * FROM titles')
print(tabulate(d, headers=["isbn", "title", "edition", "copyright"]))
```
