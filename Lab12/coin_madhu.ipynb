```python
from tkinter import *

class CoinEntries:
    Quarters = 0
    Dimes = 0
    Pennies = 0
    Nickels = 0
    Dollars = 0
    Half_Dollar = 0

def UserInput(value):
    try:
        inputValue=int(value.get())
        return inputValue
    except:
        return None
    
def Calculate():
    Cointype = CoinEntries()
    Cointype.Quarters = UserInput(Quarters1)
    if Cointype.Quarters == None or Cointype.Quarters<0:
        ErrorMsg['text'] = 'Invalid input : Quarters'
        return

    Cointype.Dimes = UserInput(Dimes1)
    if Cointype.Dimes == None or Cointype.Dimes<0:
        ErrorMsg['text'] = 'Invalid input : Dimes'
        return

    Cointype.Nickels = UserInput(Nickels1)
    if Cointype.Nickels == None or Cointype.Nickels < 0:
        ErrorMsg['text'] = 'Invalid input : Nickels'
        return

    Cointype.Pennies = UserInput(Pennies1)
    if Cointype.Pennies == None or Cointype.Pennies <0:
        ErrorMsg['text'] = 'Invalid inputs : Pennies'
        return

    Cointype.Dollars = UserInput(Dollars1)
    if Cointype.Dollars == None or Cointype.Dollars < 0:
        ErrorMsg['text'] = 'Invalid input : Dollars'
        return

    Cointype.Half_Dollar = UserInput(HalfDollar1)
    if Cointype.Half_Dollar == None or Cointype.Half_Dollar <0:
        ErrorMsg['text'] = 'Invalid inputs : Half_Dollar'
        return


    Quarters2['text']='${:.2f}'.format(Cointype.Quarters*0.25)
    Dimes2['text'] = '${:.2f}'.format(Cointype.Dimes * 0.10)
    Nickels2['text'] = '${:.2f}'.format(Cointype.Nickels * 0.05)
    Pennies2['text'] = '${:.2f}'.format(Cointype.Pennies* 0.01)
    Dollars2['text'] = '${:.2f}'.format(Cointype.Dollars* 1.00)
    HalfDollar2['text'] = '${:.2f}'.format(Cointype.Half_Dollar* 0.50)
    total=(Cointype.Dollars* 1.00)+(Cointype.Half_Dollar* 0.50)+(Cointype.Quarters*0.25)+(Cointype.Dimes*0.10)+(Cointype.Nickels*0.05)+(Cointype.Pennies*0.01)
    totalOutput['text'] = '${:.2f}'.format(total)
    ErrorMsg['text'] = ''

window=Tk()
window.title('Total Amount Calculator')
window.geometry("500x250")
window.configure(background = "light blue")
window.grid_columnconfigure(0, weight=1, uniform="abc")
window.grid_columnconfigure(1, weight=1, uniform="abc")
window.grid_columnconfigure(2, weight=1, uniform="abc")
window.grid_columnconfigure(3, weight=1, uniform="abc")
window.grid_columnconfigure(4, weight=1, uniform="abc")
window.grid_columnconfigure(5, weight=1, uniform="abc")

Label(window,text=" Enter number of coins and  \n Calculate total dollar value").grid(row=0,column=0,columnspan=4)

Label(window,text='Quarters:', width = 10).grid(row=1,column=0)
Quarters1=Entry(window, width = 10)
Quarters1.grid(row=1,column=1)


Label(window,text='Quarter value:', width = 15).grid(row=1,column=2)
Quarters2=Label(window,text='${:.2f}'.format(0.00))
Quarters2.grid(row=1,column=3)

Label(window,text='Dimes:', width = 10).grid(row=2,column=0)
Dimes1=Entry(window, width = 10)
Dimes1.grid(row=2,column=1)
Label(window,text='Dime value:', width = 15).grid(row=2,column=2)
Dimes2=Label(window,text='${:.2f}'.format(0.00))
Dimes2.grid(row=2,column=3)

Label(window,text='Nickels:', width = 10).grid(row=3,column=0)
Nickels1=Entry(window, width = 10)
Nickels1.grid(row=3,column=1)
Label(window,text='Nickel value:', width = 15).grid(row=3,column=2)
Nickels2=Label(window,text='${:.2f}'.format(0.00))
Nickels2.grid(row=3,column=3)

Label(window,text='Pennies:', width = 10).grid(row=4,column=0)
Pennies1=Entry(window, width = 10)
Pennies1.grid(row=4,column=1)
Label(window,text='Penny value:', width = 15).grid(row=4,column=2)
Pennies2=Label(window,text='${:.2f}'.format(0.00))
Pennies2.grid(row=4,column=3)

Label(window,text='Dollars:', width = 10).grid(row=5,column=0)
Dollars1=Entry(window, width = 10)
Dollars1.grid(row=5,column=1)
Label(window,text='$ value:', width = 15).grid(row=5,column=2)
Dollars2=Label(window,text='${:.2f}'.format(0.00))
Dollars2.grid(row=5,column=3)


Label(window,text='Half Dollar:', width = 10).grid(row=6,column=0)
HalfDollar1=Entry(window, width = 10)
HalfDollar1.grid(row=6,column=1)
Label(window,text='Half $ value:', width = 15).grid(row=6,column=2)
HalfDollar2=Label(window,text='${:.2f}'.format(0.00))
HalfDollar2.grid(row=6,column=3)


Button(window,text='Calculate',command= lambda : Calculate()).grid(row=7,column=0,columnspan=2)
Label(window,text='Total Amount:').grid(row=7,column=2)   
totalOutput=Label(window,text='${:.2f}'.format(0.00))
totalOutput.grid(row=7,column=3)

ErrorMsg=Label(window)
ErrorMsg.grid(row=8,column=1,columnspan=2)
window.mainloop()
```
