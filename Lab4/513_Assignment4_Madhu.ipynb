Program to plot a pie chart, line graph and bar graph using functions using matplotlib library.

Functions to plot a pie chart that shows the percentage of employees in each department within a company using provided file named employee_count_by_department.txt and use last_ten_year_net_profit.txt file to plot line and bar graph which shows companys profit over the past ten years.

```python
import matplotlib.pyplot as plt                                          # to plot graphs
from decimal import Decimal                                              # to use decimal module


def PlotPieChart(xaxis_values,yaxis_values):                             # Function to plot a Pie Chart
   piechart = plt.figure(figsize=(8,6))
   plt.pie(yaxis_values, labels = xaxis_values,autopct='%1.2f%%')        # autopct enables to display the percent value
   plt.title("Percentage of Employees per Department")
   plt.show()
    
    
def PlotLineGraph(xaxis_values,yaxis_values):                            # Function to plot a Line Graph
   linegraph = plt.figure(figsize=(7,5))
   plt.plot(xaxis_values,yaxis_values,marker="o")
   plt.grid(True)
   plt.xlim(2008, 2019)    
   plt.xlabel("Years")
   plt.ylabel("Net Profit in the 100 millions")
   plt.title("Company's Profit Over the Past Ten Years")
   plt.show()
    
    
def PlotBarGraph(xaxis_values,yaxis_values):                             # Function to plot a Bar Graph
   bargraph = plt.figure(figsize=(6,4))
   plt.bar(xaxis_values,yaxis_values)
   plt.xlim(2008, 2019)
   plt.xlabel("Years")
   plt.ylabel("Net Profit in the 100 millions")
   plt.title("Company's Profit Over the Past Ten Years")
   plt.show()   
    
    

def main():
   xaxis_values=[]                                                       # empty list for x values
   yaxis_values=[]                                                       # empty list for y values
     
   
   with open("employee_count_by_department.txt","r") as ReadFile:        # Reading existing file
       lines = ReadFile.readlines()       
       for line in lines[1:]:
           
           splitline = line.split(",")           
           xaxis_values.append(splitline[0])
           yaxis_values.append(int(splitline[1]))
   
   PlotPieChart(xaxis_values,yaxis_values)                              # Function to plot a pie chart

   xaxis_values = []                                                    # empty list for x values
   yaxis_values = []                                                    # empty list for y values
   
   with open("last_ten_year_net_profit.txt","r") as ReadFile:           # Reading existing file
       data = ReadFile.readlines()                                      # Reading line by line
       
       for line in data[1:]:                                            # For lines except first row
           
           splitline = line.split(";")                                  # Spliting lines with ; 
           xaxis_values.append(int(splitline[0]))          
           profit_amount=0
           profit_amount = Decimal("".join(num for num in splitline[1] if num.isdigit() or num == '.'))# Converting currency    
           yaxis_values.append(profit_amount)                                                            # into decimal      
  
       PlotLineGraph(xaxis_values,yaxis_values)                             # Function to plot a line graph
    
       PlotBarGraph(xaxis_values,yaxis_values)                              # Function to plot a bar graph  
   
main()

```
